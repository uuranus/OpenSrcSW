```python
#실습 1 서울의 일교차

import csv
f=open('seoul.csv','r',encoding='cp949')
data=csv.reader(f)
header=next(data)

maxdiff=0.0
maxdiff_date=''

mintemp=999.0
mintemp_date=''

maxtemp=-999.0
maxtemp_date=''

for row in data:
    if(row[-1]==''): #비어있는 정보
        continue
    if(float(row[-1])-float(row[-2])>=maxdiff):
        maxdiff=float(row[-1])-float(row[-2])
        maxdiff_date=row[0]
    
    if(float(row[-1])>=maxtemp):
        maxtemp=float(row[-1])
        maxtemp_date=row[0]
        
    if(float(row[-2])<=mintemp):
        mintemp=float(row[-2])
        mintemp_date=row[0]
        
f.close()
print('일교차가 가장 큰 날')
print(maxdiff_date)
print('최저기온이 가장 낮은 날')
print(mintemp_date)
print('최고기온이 가장 높은 날')
print(maxtemp_date)
```

```python
#실습 2 지하철 이용객

import csv
f=open('subway.csv','r',encoding='UTF-8')
data=csv.reader(f)
header=next(data)

before_station_line=''
station_line=''
use_ranking={}

for row in data:
    station_line=row[1]
    if(station_line!=before_station_line): #호선이 바뀌었다면
        if(before_station_line!=''): #1호선이 아니라면
            use_ranking=sorted(use_ranking.items(),reverse=True,key=lambda x:x[1]) #내림차순 정렬
            #정렬하고 나면 튜플 집합 list로 변경됨
            print(before_station_line)
            print('이용객이 가장 많은 역 순위')
            print("1. {0} ({1}) ".format(use_ranking[0][0],use_ranking[0][1]))
            print("2. {0} ({1}) ".format(use_ranking[1][0],use_ranking[1][1]))
            print("3. {0} ({1}) ".format(use_ranking[2][0],use_ranking[2][1]))
            print()
            print('이용객이 가장 적은 역 순위')
            print("1. {0} ({1}) ".format(use_ranking[-1][0],use_ranking[-1][1]))
            print("2. {0} ({1}) ".format(use_ranking[-2][0],use_ranking[-2][1]))
            print("3. {0} ({1}) ".format(use_ranking[-3][0],use_ranking[-3][1]))
            print()
            
        before_station_line=station_line
        use_ranking={}
        
        #row[-1]은 ,
        on=row[-3].split(',')
        newon=''
        for i in range(len(on)):
            newon+=on[i]
            
        on=int(newon)
               
        off=row[-2].split(',')
        newoff=''
        for i in range(len(off)):
            newoff+=off[i]
            
        off=int(newoff)
                   
        use_ranking[row[-4]]=on+off
    else:
        #row[-1]은 ,
        on=row[-3].split(',')
        newon=''
        for i in range(len(on)):
            newon+=on[i]
            
        on=int(newon)
               
        off=row[-2].split(',')
        newoff=''
        for i in range(len(off)):
            newoff+=off[i]
            
        off=int(newoff)
                   
        use_ranking[row[-4]]=on+off
        
#마지막 호선 출력
use_ranking=sorted(use_ranking.items(),reverse=True,key=lambda x:x[1]) #내림차순 정렬
#정렬하고 나면 튜플 집합 list로 변경됨
print(before_station_line)
print('이용객이 가장 많은 역 순위')
print("1. {0} ({1}) ".format(use_ranking[0][0],use_ranking[0][1]))
print("2. {0} ({1}) ".format(use_ranking[1][0],use_ranking[1][1]))
print("3. {0} ({1}) ".format(use_ranking[2][0],use_ranking[2][1]))
print()
print('이용객이 가장 적은 역 순위')
print("1. {0} ({1}) ".format(use_ranking[-1][0],use_ranking[-1][1]))
print("2. {0} ({1}) ".format(use_ranking[-2][0],use_ranking[-2][1]))
print("3. {0} ({1}) ".format(use_ranking[-3][0],use_ranking[-3][1]))
    
```
