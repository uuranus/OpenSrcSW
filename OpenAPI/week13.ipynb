```python
#실습 1 생일 기온 시각화
import matplotlib.pyplot as plt
import csv

f=open('Temp10.csv')
data=csv.reader(f)
header=next(data)

result=[]
for row in data:
    if(len(row)==0):
        continue
    if row[-1]=='':
        continue
    else:
        data=row[0].split('-')
        if(data[1]=='09' and data[2]=='20'): #내 생일날이면
            result.append(float(row[-1])) #그날의 최고기온을 저장
                          
plt.title('highest temperature')
plt.plot(result)
plt.show()
```

```python
#실습 2 최고 기온 시각화

import matplotlib.pyplot as plt
import csv

f=open('Temp10.csv')
data=csv.reader(f)
header=next(data)

Jan=[] #1월의 데이터
Aug=[] #8월의 데이터

for row in data:
    if len(row)==0:
        continue
    if row[-1]=='':
        continue
    
    data=row[0].split('-')
    if(data[1]=='01'): #1월이면
        Jan.append(float(row[-1])) #그날의 최고기온을 저장
        
       
    elif(data[1]=='08'): #8월이면
        Aug.append(float(row[-1])) #그날의 최고기온을 저장
        
        
plt.title('higest temperature')
plt.hist(Jan,bins=100,label='January')
plt.hist(Aug,bins=100,label='August')
plt.legend()
plt.show()
```
